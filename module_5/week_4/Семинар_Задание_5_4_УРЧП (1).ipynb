##Теоретическая часть.##

Для простейшего уравнения переноса

$$
u_t+\lambda u_x=0, \quad \lambda=\text { const } \neq 0
$$

$(d x=\lambda d t \quad-$ характеристика уравнения (1)), все множество явных линейных схем $u_m^{n+1}=\sum_{\mu, v} \alpha_\mu^v(\tau, h) u_{m+\mu}^{n+v}$ на 5-точечном сеточном шаблоне $\left(t^{n+1}, x_m\right),\left(t^n, x_{m-2}\right),\left(t^n, x_{m-1}\right),\left(t^n, x_m\right)$, $\left(t^n, x_{m+1}\right),\left(t^n, x_{m+2}\right)$ (рис. 1): 

image.png

с 5-ю неопределенными коэффициентами $\left\{\alpha_{-2}^0, \alpha_{-1}^0, \alpha_0^0, \alpha_1^0, \alpha_2^0\right\}$ имеет вид:
$$
u_m^{n+1}=\alpha_{-2}^0 u_{m-2}^n+\alpha_{-1}^0 u_{m-1}^n+\alpha_0^0 u_m^n+\alpha_1^0 u_{m+1}^n+\alpha_2^0 u_{m+2}^n .
$$


**Определение.**

Однородные разностные схемы, сохраняющие монотонность пространственного профиля разностного решения, называются монотонными.

**Теорема Фридрихса.**

Явная двухслойная линейная однородная схема, аппроксимирующая эволюционно уравнение и записанная в виде
$$
u_m^{n+1}=\sum_{\mu, v} \alpha_\mu^v(\tau, h) u_{m+\mu}^{n+v}
$$
монотонна тогда и только тогда, когда все коэффициенты > 0 .
Такие разностные схемы называются **монотонными по Фридрихсу**.


**Теорема Годунова.**

Среди линейных разностных схем для уравнения переноса с порядком аппроксимации выше первого нет монотонных



После удовлетворения условий аппроксимации первого порядка:
$$
\delta_0=-1+\sum_{\mu, v} \alpha_\mu^v=0, \delta_1=\sigma+\sum_{\mu, v}(\mu-\sigma v) \alpha_\mu^v=0
$$  

( $\sigma=\lambda \tau / h$ - число Куранта) три коэффициента, например, $\alpha_{-2}^0, \alpha_0^0$ и $\alpha_2^0$, остаются свободными, а два других, например, $\alpha_{-1}^0, \alpha_1^0$ и, выражаются через эти свободные коэффициенты. Принимая $\alpha_{-2}^0, \alpha_0^0$ и $\alpha_2^0$ за координаты линейного пространства с евклидовой метрикой, в котором каждой точке соответствует некоторая схема 1-го порядка аппроксимации, можно найти все множество устойчивых схем (при $0<\sigma<1$ и $\alpha_2^0=0$ - область $B_1, B_2, C, B_3, B_4, A_3, A_4, A_0$ на рис. 2), все монотонные схемы, у которых все коэффициенты $\alpha_\mu^v \geq 0$, (при $0<\sigma<1$ и $\alpha_2^0=0$ - четырехугольник $A_0, A_1, A_3, A_4$ на рис. 2), двухпараметрическое семейство схем 2-го порядка аппроксимации
$$
\delta_2=-\sigma^2+\sum_{\mu, v}(\mu-\sigma v)^2 \alpha_\mu^v p_\mu^v=0
$$ 

(отрезок $B_1-B_4$ на рис. 2 - устойчивые при $0<\sigma<1$ и $\alpha_2^0=0$ схемы с порядком аппроксимации выше первого), однопараметрическое семейство схем 3-го порядка аппроксимации:
$$
\delta_3=\sigma^3+\sum_{\mu, v}(\mu-\sigma v)^3 \alpha_\mu^v q_\mu^v=0
$$
(при $0<\sigma<1$ и $\alpha_2^0=0$ точка $C$ на рис. 2 ) и единственную на этом шаблоне схему 4-го порядка:
$$
\delta_4=-\sigma^4+\sum_{\mu, v}(\mu-\sigma v)^4 \alpha_\mu^v r_\mu^v=0 .
$$

Здесь $p_\mu^v$,  $q_\mu^v$,  $r_\mu^v$ - коэффициенты, которые можно из ряда Тейлора найти. 

Также можно получать оптимальные в том или ином смысле схемы, например, «наиболее точную из монотонных схем» - точка $A_1$, «наименее осциллирующую на разрывных решениях  

схему 2-го порядка аппроксимации» - точка $B_2$ на рис. 2 , или монотонные схемы с переменным порядком аппроксимации - вторым или третьим в зависимости от поведения решения и т.д.  

**На примере нашего шаблона**

\end{aligned} 


\begin{aligned}

Собираем коэффициенты, чтобы левая часть была равна правой.

В нулевом порядке аппроксимации 

$$
\alpha_{-2}^0+\alpha_{-1}^0+\alpha_0^0+\alpha_1^0+\alpha_2^0=1  
$$

В первом порядке аппроксимации 
$$
\frac{-\lambda \tau}{h}=-2 \alpha_{-2}^0-\alpha_{-1}^0+\alpha_1^0+2 \alpha_2^0 
$$ 

Вводим обозначение числа Куранта
$$
\sigma=\frac{\lambda \tau}{h}
$$

Во втором порядке аппроксимации

\begin{gathered}
\left(u_m^n\right)_{t t}^{\prime \prime}=(-\lambda)^2 \cdot\left(u_m^n\right)_{x x}^{\prime \prime}=\lambda^2\left(u_m^n\right)_{x x}^{\prime \prime} \\
\frac{1}{2} \sigma^2=2 \alpha_{-2}^0+\frac{1}{2} \alpha_{-1}^0+\frac{1}{2} \alpha_1^0+2 \alpha_2^0
\end{gathered}

В третьем порядке аппроксимации

\begin{aligned}
& \left(u_m^n\right)_{t t t}^{\prime \prime \prime}=(-\lambda)^3 \cdot\left(u_m^n\right)_{x x x}^{\prime \prime \prime}=-\lambda^3\left(u_m^n\right)_{x x x}^{\prime \prime \prime} \\
& -\frac{\sigma^3}{6}=-\frac{4}{3} \alpha_{-2}^0-\frac{1}{6} \alpha_{-1}^0+\frac{1}{6} \alpha_1^0+\frac{4}{3} \alpha_2^0
\end{aligned}  

В четвертом порядке аппроксимации
$$
\frac{\sigma^4}{24}=\frac{2}{3} \alpha_{-2}^0+\frac{1}{24} \alpha_{-1}^0+\frac{1}{24} \alpha_1^0+\frac{2}{3} \alpha_2^0  
$$

Оставим свободными три коэффициента и выразим оставшиеся два из условий аппроксимации нулевого и первого порядка

\begin{aligned}

\alpha_{-1}^0+\alpha_1^0=1-\alpha_{-2}^0-\alpha_2^0-\alpha_0^0 \\
\alpha_1^0-\alpha_{-1}^0=-\sigma+2 \alpha_{-2}^0-2 \alpha_2^0
\end{array}\right. \\
& \alpha_1^0=\frac{1}{2}(1-\sigma)+\frac{1}{2}\left(\alpha_{-2}^0-3 \alpha_2^0-\alpha_0^0\right) \\
& \alpha_{-1}^0=\frac{1}{2}(1+\sigma)+\frac{1}{2}\left(\alpha_2^0-3 \alpha_{-2}^0-\alpha_0^0\right)
\end{aligned}

Берем 
$$
\text { при } 0<\sigma<1 \ \ \  \alpha_2^0=0 \\
\text { при }-1<\sigma<0 \ \ \ \  \alpha_{-2}^0=0
$$

При $-1<\sigma<0$ и $\alpha_{-2}^0=0$ аналогичные точкам $A_0-C$ на рис. 2 разностные схемы (т.е. точки $A_0, A_1^{\prime}-A_4^{\prime}, B_1^{\prime}-B_6^{\prime}, C$ на рис. 2a) получаются заменой $\sigma$ на $-\sigma$ и коэффициентов $\alpha_\mu^v$ на $\alpha_{-\mu}^v$. Например, если для т. $A_1$ (при $\sigma>0$ ) имеем:
$$
\alpha_{-2}^0=\alpha_2^0=\alpha_1^0=0, \alpha_{-1}^0=\sigma, \alpha_0^0=1-\sigma
$$
то для т. $A_1^{\prime}($ при $\sigma<0)$ :
$$
\alpha_{-2}^0=\alpha_2^0=\alpha_{-1}^0=0, \alpha_1^0=-\sigma, \alpha_0^0=1+\sigma
$$
Для построения монотонной схемы со вторым-третьим порядком аппроксимации каждый из коэффициентов $\alpha_\mu^v$ вектора $\vec{\alpha}=\left\{\alpha_{-2}^0, \ldots, \alpha_2^0\right\}$ можно задавать в виде линейной комбинации трех опорных схем - точек в пространстве неопределенных коэффициентов (т.е. одной схемы 3го порядка т. $C-C^{\prime}$ и двух схем 2-го порядка аппроксимации т. $B_3, B_5-B_3^{\prime}, B_5^{\prime}$ с ортогональными областями монотонности). Выбор между схемами при этом делать так, чтобы удовлетворялся характеристический критерий монотонности:
$$
\begin{aligned}
& \left\{\begin{array}{l}
u_{m-1}^n<u_m^{n+1}=u_*<u_m^n, \text { если } \Delta_0=u_{m-1}^n-u_m^n<0 \\
u_{m-1}^n>u_m^{n+1}=u_*>u_m^n, \text { если } \Delta_0=u_{m-1}^n-u_m^n>0
\end{array}, \sigma>0\right. \\
& \left\{\begin{array}{l}
u_m^n<u_m^{n+1}=u_*<u_{m+1}^n, \text { если } \Delta_0=u_{m+1}^n-u_m^n>0 \\
u_m^n>u_m^{n+1}=u_*>u_{m+1}^n, \text { если } \Delta_0=u_{m+1}^n-u_m^n<0
\end{array}, \sigma<0\right. \\
&
\end{aligned}
$$
который может быть записан с учетом условий аппроксимации второго порядка (4) в следующем виде: 

\begin{aligned}
& \left\{\begin{array}{l}
\Delta_0<\sigma\left(\Delta_0+\Delta_1\right) / 2+\sigma^2\left(\Delta_0-\Delta_1\right) / 2+\alpha_{-2}^0\left(\Delta_{-1}-2 \Delta_0+\Delta_1\right)<0, \text { если } \Delta_0<0 \\
\Delta_0>\sigma\left(\Delta_0+\Delta_1\right) / 2+\sigma^2\left(\Delta_0-\Delta_1\right) / 2+\alpha_{-2}^0\left(\Delta_{-1}-2 \Delta_0+\Delta_1\right)>0, \text { если } \Delta_0>0 \\
\sigma>0, \quad \alpha_2^0=0, \Delta_{-1}=u_{m-2}^n-u_{m-1}^n, \Delta_0=u_{m-1}^n-u_m^n, \Delta_1=u_m^n-u_{m+1}^n
\end{array}\right. \\
& \left\{\begin{array}{l}
0<\sigma\left(\Delta_0+\Delta_1\right) / 2+\sigma^2\left(\Delta_0-\Delta_1\right) / 2+\alpha_2^0\left(\Delta_{-1}-2 \Delta_0+\Delta_1\right)<\Delta_0, \text { если } \Delta_0>0 \\
0>\sigma\left(\Delta_0+\Delta_1\right) / 2+\sigma^2\left(\Delta_0-\Delta_1\right) / 2+\alpha_2^0\left(\Delta_{-1}-2 \Delta_0+\Delta_1\right)>\Delta_0, \text { если } \Delta_0<0 \\
\sigma<0, \quad \alpha_{-2}^0=0, \Delta_{-1}=u_{m+2}^n-u_{m+1}^n, \Delta_0=u_{m+1}^n-u_m^n, \Delta_1=u_m^n-u_{m-1}^n
\end{array}\right.
\end{aligned}  


Разделив затем оба неравенства на $\Delta_0$, получаем характеристическое условие монотонности в свернутом виде:
$$
\begin{aligned}
& \left\{\begin{array}{l}
0 \leq \sigma\left(1+\delta_1\right) / 2+\sigma^2\left(1-\delta_1\right) / 2+\alpha_{-2}^0\left(\delta_{-1}-2+\delta_1\right) \leq 1 \\
\sigma>0, \quad \alpha_2^0=0, \quad \delta_{-1}=\Delta_{-1} / \Delta_0, \quad \delta_1=\Delta_1 / \Delta_0, \quad \Delta_0 \neq 0
\end{array}\right. \\
& \left\{\begin{array}{l}
0 \leq|\sigma|\left(1+\delta_1\right) / 2+\sigma^2\left(1-\delta_1\right) / 2+\alpha_2^0\left(\delta_{-1}-2+\delta_1\right) \leq 1 \\
\sigma<0, \quad \alpha_{-2}^0=0, \quad \delta_{-1}=\Delta_{-1} / \Delta_0, \quad \delta_1=\Delta_1 / \Delta_0, \quad \Delta_0 \neq 0
\end{array}\right.
\end{aligned}
$$
Схемы 2-го порядка аппроксимации с ортогональными областями монотонности (точки $B_3, B_5-B_3^{\prime}, B_5^{\prime}$ для демонстрационного шаблона) можно легко найти, подбирая так значения $\alpha_{-2}^0$ (при $\left.\sigma>0\right), \alpha_2^0$ (при $\sigma<0$ ) в неравенствах (9), чтобы в них оставалась зависимость только от одного из параметров $\delta_{-1}$ или $\delta_1$. В нашем случае это легко увидеть, подставив координаты точек $\quad B_3 \rightarrow\left\{\alpha_{-2}^0=0\right\}, \quad B_5 \rightarrow\left\{\alpha_{-2}^0=\sigma(\sigma-1) / 2\right\} \quad$ (при $\left.\quad \sigma>0\right) \quad$ или $\quad B_3^{\prime} \rightarrow\left\{\alpha_2^0=0\right\}$, $B_5^{\prime} \rightarrow\left\{\alpha_2^0=\left(\sigma^2-|\sigma|\right) / 2\right\}($ при $\sigma<0)$ в (9).  

**На примере нашего шаблона** 

Получим аналитический вид  для однопараметрического множества схем второго порядка.

$$
\frac{1}{2} \sigma^2=2 \alpha_{-2}^0+\frac{1}{2} \alpha_{-1}^0+\frac{1}{2} \alpha_1^0
$$

\begin{aligned}
& \frac{\sigma^2}{2}=2 \alpha_{-2}^0+\frac{1}{2}\left(1+\left(-\alpha_0^0-\alpha_{-2}^0\right)\right) \\
& \sigma^2=4 \alpha_{-2}^0+1-\alpha_0^0-\alpha_{-2}^0 \\
& \sigma^2=3 \alpha_{-2}^0-\alpha_0^0+1 \\
& \alpha_0^0-3 \alpha_{-2}^0=1-\sigma^2 \\
& \alpha_0^0=1-\sigma^2+3 \alpha_{-2}^0
\end{aligned}

Это уравнение прямой на плоскости параметров. 

Найдем схему 3-го порядка

\begin{aligned}
& -\frac{\sigma^3}{6}=-\frac{4}{3} \alpha_{-2}^0-\frac{1}{6} \alpha_{-1}^0+\frac{1}{6} \alpha_1^0 \\
& \sigma^3=8 \alpha_{-2}^0+\alpha_{-1}^0-\alpha_1^0 \\
& \sigma^3=8 \alpha_{-2}^0+\frac{1}{2}\left(1+\sigma\right)-\frac{3}{2} \alpha_{-2}^0-\frac{1}{2} \alpha_0^0 \\
& +\frac{1}{2}(\sigma-1)-\frac{1}{2} \alpha_{-2}^0+\frac{1}{2} \alpha_0^0 \\
& \sigma^3=6 \alpha_{-2}^0+\sigma \\
& \alpha_{-2}^0=\frac{\sigma^3-\sigma}{6} ;\  \alpha_0^0=1-\sigma^2+3 \alpha_{-2}^0\\
& \alpha_1^0=\frac{1}{2}(1-\sigma)+\frac{\sigma^3-\sigma}{12}-\frac{\alpha_0^0}{2} \\
& \alpha_{-1}^0=\frac{1}{2}(1+\sigma)-\frac{\sigma^3-\sigma}{4}-\frac{\alpha_0^0}{2}
\end{aligned}  


Доведя вычисления до конца, получаем

\begin{aligned}
& \alpha_0^0=\frac{\sigma^3}{2}-\sigma^2-\frac{\sigma}{2}+1= \\
& =\left(\frac{\sigma}{2}-1\right) \cdot\left(\sigma^2-1\right) \\
& \alpha_1^0=-\frac{\sigma^3}{6}+\frac{\sigma^2}{2}-\frac{\sigma}{3} \\
& \alpha_{-1}^0=-\frac{\sigma^3}{2}+\frac{\sigma^2}{2}+\sigma \\
& \alpha_{-2}^0=\frac{\sigma^3-\sigma}{6}
\end{aligned}
