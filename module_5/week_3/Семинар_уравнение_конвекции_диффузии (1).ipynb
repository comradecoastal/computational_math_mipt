Требуется решить двумерное уравнение конвекции-диффузии: 

$u_t+\lambda_1 u_x+\lambda_2 u_y=\kappa\left(u_{x x}+u_{y y}\right)+f$ 

с использованием простейшей разностной схемы типа "крест» и равномерной расчетной сетки $\Delta \mathrm{x}=\Delta y=h$. 

В случае стационарных граничных условий $\left.u\right|_{\mathrm{r}}=\varphi(x, y)$ pешение данного уравнения будет стремиться к решению уравнения

 $\lambda_1 u_x+\lambda_2 u_y=\kappa\left(u_{x x}+u_{y y}\right)+f$ 
 
 с теми же граничными условиями 
 
 $\left.u\right|_{\mathrm{r}}=\varphi(x, y)$. 
 
 Здесь $u$ - удельная концентрация переносимого вещества, выраженная в $мг/{ м}^3, $
 
 $\lambda_1, \lambda_2$ - проекции вектора скорости переноса вещества на оси $x, y$,  в пределах от 1 до 10 $м/с$.
 
 $ \kappa>0$ - эффективный коэффициент диффузии
 
 правая часть - $f$ позволяет моделировать источники или наоборот, стоки переносимого вещества.


\begin{aligned}
& U(t, x, y) \rightarrow U_{m n}^k \\
& 0 \leqslant m \leqslant M, h_x=h \\
& 0 \leqslant n \leqslant N, h_y=h \\
& 0 \leqslant k \leqslant \frac{T}{T} ; \Delta t=\tau
\end{aligned}


Из лекции необходимые условия устойчивости схемы

\begin{aligned}
& h \leq \frac{6 \kappa}{|\lambda_{\text {max }} \mid} \\
& \tau<\frac{h}{\mid \lambda_{\text {max }}|}
\end{aligned} 

Метод переменных направлений позволяет быстро и эффективно решить задачу. 

Он заключается в том, что мы делим шаг по времени на два полушага, на одном из них на новом слое пишем производную только по х, на другом - только по у.

$$
L_x\left(U_{m n}^k\right)=\frac{U_{m+1, n}^k-2 \cdot U_{m , n}^k+U_{m-1, n}^k}{h^2} 
$$


$$
L_y\left(U_{m n}^k\right)=\frac{U_{m, n+1}^k-2 \cdot U_{m , n}^k+U_{m, n-1}^k}{h^2}

Полные формулы метода:

Второй полушаг
