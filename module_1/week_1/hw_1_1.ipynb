```python
# 8

n = 50
time_numpy = np.zeros(n)
time_loop = np.zeros(n)

def dumb_matrix_prod(a, b, n):
  result = np.zeros((n, n))
  for i in range(n):
    for j in range(n):
      prod = 0
      for k in range(n):
        prod += a[i, k] * b[k, j]
      result[i, j] = prod
  return result

# проверка 
# a = sps.randint.rvs(0, 10, size=(3, 3))
# b = sps.randint.rvs(0, 10, size=(3, 3))
# print(a, '\n', b)
# print(a@b)
# print(dumb_matrix_prod(a, b, 3))

for i in range(1, n + 1):
  a = sps.uniform.rvs(size=(i, i), loc=0, scale=10)
  b = sps.uniform.rvs(size=(i, i), loc=0, scale=10)
  start = time.perf_counter()
  a @ b
  time_numpy[i - 1] = time.perf_counter() - start
  start = time.perf_counter()
  dumb_matrix_prod(a, b, i)
  time_loop[i - 1] = time.perf_counter() - start

plt.figure(figsize=(10, 7))
plt.subplot(211)
plt.grid(True)
plt.title("Время для numpy")
plt.plot(np.arange(1, n + 1), time_numpy)
plt.subplot(212)
plt.grid(True)
plt.title("Время для цикла for")
plt.plot(np.arange(1, n + 1), time_loop)
```

### Упражнение 3 Вигнеровский полукруг

```python
# 1

n = 1000
distr = sps.norm(loc=0, scale=1)
eigenvalues = np.array([])

for _ in range(10):
    m = distr.rvs(size=(n, n))
    m += m.T
    eigenvalues = np.append(eigenvalues, np.linalg.eigvals(m))
    
plt.hist(eigenvalues, bins=50, density=True)
plt.show()
```
