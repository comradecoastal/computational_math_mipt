```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

случайная симметричная матрица

```python
A = np.random.uniform(0, 1, size=(3, 3))
A = A @ A.T / 2
print(A)
```

```python
N = 200
X = np.random.multivariate_normal([0, 0, 0], A, size=N)
```

```python
fig = plt.figure(figsize=(15, 10))
ax = plt.axes(projection="3d")

ax.scatter(X[:, 0], X[:, 1], X[:, 2])

pca = PCA(n_components=3)
pca.fit(X)

colors = ("red", "green", "blue")

for i in range(3):
    x, y, z = pca.components_[i]
    ax.plot([0, x], [0, y], [0, z], color=colors[i], label=f"Сингулярный вектрор {i}")


plt.legend()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$x$")
    
plt.show()
```

Проекции на 3-х графиках

```python
data = pca.transform(X)
fig, axs = plt.subplots(1, 3, figsize=(21, 7))
print(data.shape)

axs[0].scatter(data[:, 0], data[:, 1])
axs[0].set_title("Проекция на 1 и 2 синг. вектора")
axs[1].scatter(data[:, 1], data[:, 2])
axs[1].set_title("Проекция на 2 и 3 синг. вектора")
axs[2].scatter(data[:, 2], data[:, 0])
axs[2].set_title("Проекция на 3 и 1 синг. вектора")

plt.show()
```
